```python
import pandas as pd
import numpy as np
import random
import string
import math
```

```python
df = pd.read_csv('../Y2015-T3Y2023.csv')

#DimTable Read
df_Airport_Ref = pd.read_excel('20230409-AirportInfo(2015-2022)-UniPage.xlsx')
df_Airport_Vietnam = pd.read_excel('20230410-AirportInVietnam.xlsx')
df_Airline_Ref = pd.read_excel('IATA_Airline_Info(2017-2022).xlsx')
df_Revenue_Report = pd.read_excel('DT theo CN (R).xlsx')
```

```python
df['FlightDate'] = pd.to_datetime(df['FlightDate'])
```

```python
#Time duration condition
df = df[df['FlightDate'] <= '2022-12-31']
```

```python
df.shape
```

```python
df = df[~((df['Airline'] == '**') | (df['Airline'] == '##'))]
df = df[df['Route'].str.len() == 7]
```

```python
remove_n = math.floor(0.1*df.shape[0])
drop_indices = np.random.choice(df.index, remove_n, replace=False)
df = df.drop(drop_indices).reset_index(drop = True)
```

```python
df.shape
```

### <span style = "color:red">**Random Airports** </span>

```python
# set1 = set(df['Route'].str[:3])
# set2 = set(df['Route'].str[4:])
# AirportList = list(set(set1) | set(set2))

AirportList = list(df_Airport_Ref['Airport Code'].unique())
```

```python
df['Origin'] = df['Route'].str[:3]
df['Destination'] = df['Route'].str[4:]
```

```python
dct = {}
for p in AirportList:
    x = ''.join(random.SystemRandom().choice(string.ascii_uppercase) for _ in range(3)) #change the number for the following result
    dct[p] = x
```

```python
df['Origin'] = df['Origin'].map(dct)
df['Destination'] = df['Destination'].map(dct)
df['BaseStation'] = df['BaseStation'].map(dct)

df['Route'] = df['Origin']  + '-' + df['Destination']
#DimTable Change
df_Airport_Ref['Airport Code'] = df_Airport_Ref['Airport Code'].map(dct)
df_Airport_Vietnam['Airport_Code'] = df_Airport_Vietnam['Airport_Code'].map(dct)
df_Revenue_Report['CHK'] = df_Revenue_Report['CHK'].map(dct)
```

```python
df_Airport_Vietnam
```

### <span style = "color:red">**Random Airline** </span>

```python
# AirlineList = list(df['Airline'].unique())


AirlineList = list(df_Airline_Ref['Airline Code'].unique())
```

```python
dct2 = {}
for p in AirlineList:
    x = ''.join(random.SystemRandom().choice(string.ascii_uppercase) for _ in range(2)) 
    dct2[p] = x
```

```python
df_Airline_Ref['Airline Code'] = df_Airline_Ref['Airline Code'].map(dct2)
```

```python
df['Airline'] = df['Airline'].map(dct2)
```

```python
df['Num'] = df['FlightNo'].str[2:]
```

```python
FlightNumList = list(df['Num'].unique())
```

```python
dct3 = {}
for p in FlightNumList:
    numchar = str(p)
    x = ''.join(random.SystemRandom().choice(string.digits) for _ in range(len(numchar))) 
    dct3[p] = x
```

```python
df['Num'] = df['Num'].map(dct3)
```

```python
df['FlightNo'] = df['Airline'] + df['Num']
```

```python
df.drop(['Num', 'Origin', 'Destination'], axis = 1, inplace = True)
df
```

```python
import os
isExist = os.path.exists('ProjectData')
if not isExist:
    os.mkdir('ProjectData')
```
